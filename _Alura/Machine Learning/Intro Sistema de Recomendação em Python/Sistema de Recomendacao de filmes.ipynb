## Algoritmo para recomendação de filmes

```python
## importando bibliotecas

import pandas as pd
```

```python
## importando dataset filmes
filmes = pd.read_csv('movies.csv')
filmes.columns=['filmeId', 'titulo', 'generos'] ##alterando o nome das colunas para pt
filmes = filmes.set_index('filmeId') ## transformando a variavel 'filmeId' em indice
filmes.head()
```

```python
## Importando o dataset de notas
notas = pd.read_csv('ratings.csv')
notas.columns =['usuarioid', 'filmeid', 'nota', 'momento'] ## alterando o nome das colunas para pt
notas.head()
```

## Recomendação de filmes mais populares

```python
## Criando a variavel com calculo da media de notas dos filmes mais votados

total_de_votos = notas['filmeid'].value_counts()
```

```python
## Consultando qual foi o filme com indice 356 (apenas teste)
filmes.loc[356] 
```

```python
##Incluindo a quantidade de votos dos filmes no dataframe 'filmes'
filmes['total_de_votos'] = total_de_votos
filmes.head(3)
```

```python
filmes.sort_values('total_de_votos', ascending = False).head(10)
```

## Recomendação de filmes com maiores notas média

```python
## Calculando a nota media

notas_media = notas.groupby('filmeid').mean()['nota']
```

```python
## Incluindo a coluna 'notas_media' no DataFrame 'filmes'
filmes['notas_media'] = notas_media
filmes.sort_values('notas_media').head(3)
```

```python
## Filtrando os valores > 50
filmes_com_mais_50_votos = filmes.query('total_de_votos > 50').sort_values('notas_media', 
                                                                           ascending = False)

filmes_com_mais_50_votos.head(5)
```
