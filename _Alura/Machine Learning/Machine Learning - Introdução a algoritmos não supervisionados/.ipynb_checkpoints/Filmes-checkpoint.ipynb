```python
import pandas as pd
```

```python
uri_filmes = 'https://raw.githubusercontent.com/alura-cursos/machine-learning-algoritmos-nao-supervisionados/master/movies.csv'

filmes = pd.read_csv(uri_filmes)

filmes.columns = ['filme_id', 'titulo', 'generos']
filmes
```

```python
filmes.generos
```

```python
generos = filmes.generos.str.get_dummies()
```

```python
## Concatenando os dois DF

dados_dos_filmes = pd.concat([filmes, generos], axis = 1)
dados_dos_filmes.head()
```

```python
## escalar dados

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
generos_escalados = scaler.fit_transform(generos)
generos_escalados
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=3) ## Quantidade de grupos

modelo.fit(generos_escalados) ## Treinar o modelo
print('Grupos {}'.format(modelo.labels_))
```

```python
print(generos.columns)
print(modelo.cluster_centers_)
```

```python
grupos = pd.DataFrame(modelo.cluster_centers_, columns = generos.columns)
grupos
```

```python
grupos.transpose().plot.bar(subplots = True, figsize = (25, 25), sharex=False)
```

```python
grupo = 2

filtro = modelo.labels_ == grupo

dados_dos_filmes[filtro].sample(10)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE()

visulizacao = tsne.fit_transform(generos_escalados)
```

```python
visulizacao
```

```python
import seaborn as sns

sns.set(rc={'figure.figsize' : (13, 8)})
sns.scatterplot(x=visulizacao[:, 0], y=visulizacao[:, 1], hue=modelo.labels_, 
                palette=sns.color_palette('Set1', 3))
```

```python
modelo = KMeans(n_clusters = 20)
modelo.fit(generos_escalados)

grupos = pd.DataFrame(modelo.cluster_centers_, columns=generos.columns)

grupos.transpose().plot.bar(subplots = True,
                           figsize = (25, 50),
                           sharex = False,
                           rot=0)
```

```python
grupo = 3

filtro = modelo.labels_ == grupo
dados_dos_filmes[filtro].sample(10)
```

```python
def kmeans(numero_de_clusters, generos):
    modelo = KMeans(n_clusters=numero_de_clusters)
    modelo.fit(generos)
    return[numero_de_clusters, modelo.inertia_]
```

```python
kmeans(20, generos_escalados)
```

```python
kmeans(3, generos_escalados)
```

```python
resultado = [kmeans(numero_de_grupos, generos_escalados) for numero_de_grupos in range(1, 41)]
resultado
```

```python
resultado = pd.DataFrame(resultado, columns=['grupos', 'inertia'])
resultado
```

```python
resultado.inertia.plot(xticks=resultado.grupos)
```

```python
modelo = KMeans(n_clusters = 17)
modelo.fit(generos_escalados)

grupos = pd.DataFrame(modelo.cluster_centers_, columns=generos.columns)

grupos.transpose().plot.bar(subplots = True,
                           figsize = (25, 50),
                           sharex = False,
                           rot=0)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
modelo = KMeans(n_clusters = 17)
modelo.fit(generos_escalados)

grupos = pd.DataFrame(modelo.cluster_centers_, columns=generos.columns)

grupos.transpose().plot.bar(subplots = True,
                           figsize = (25, 50),
                           sharex = False,
                           rot=0)
```

```python
matriz_de_distancia = linkage(grupos)
matriz_de_distancia
```

```python
dendrograma = dendrogram(matriz_de_distancia)
```
