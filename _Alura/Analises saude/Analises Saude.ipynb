```python
## Lendo o arquivo de populacao estimada da United Nations

un_population = pd.read_csv('https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2019_TotalPopulationBySex.csv')

## Foi verificado que o dataset estava em milhares. 'Corrigindo' a variavel para que os valores sejam demonstrados em milhares
un_population['PopTotal'] = un_population['PopTotal'] * 1000

un_population.head()
```

```python
## Descrevendo o dataset
un_population.describe()
```

```python
## Fazendo a query somente com as colunas de tempo e populacao total

un_population_total = un_population.query('Time==2019')[['Location', 'PopTotal']].set_index('Location')
un_population.head()
```

```python
## Trazendo codigos do curso anterior (outro notebook)

deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")

display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
    return data.groupby("Country/Region").sum().iloc[:,-1]

display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

```python
confirmed.head()
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ["confirmed", "deaths", "recovered"]
combined.head()
```

```python
def rename_location(location):
    if location=='China':
        return 'China'
    if location=='Republic of Korea':
        return 'South Korea'
    if location=='Iran (Islamic Republic of)':
        return 'Iran'
    if location=='United States of America':
        return 'US'
    if location=='China, Hong Kong SAR':
        return 'Hong Kong'
    return location

un_population_total['location_for_who'] = un_population_total.index.map(rename_location)
un_population_total.query('location_for_who == "South Korea"').head()
```

```python
## Juntar as duas variaveis, alterando o indice para o novo indice == location_for_who

un_population_total = un_population_total.set_index('location_for_who')
un_population_total.head()
```

```python
combined_expanded = combined.join(un_population_total).sort_values('confirmed', ascending=False).dropna()
```

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 100000
```

```python
hubei_population = 59020000
confirmed_cases = confirmed.set_index('Province/State').loc['Hubei']['7/10/20']
death_cases = deaths.set_index('Province/State').loc['Hubei']['7/10/20']
recovered_cases = recovered.set_index('Province/State').loc['Hubei']['7/10/20']

incident_ratio = confirmed_cases / hubei_population * 1000000
mortality_rate = death_cases / hubei_population * 1000000
```

```python
print(f'Coeficiente de incidencia em Hubei {incident_ratio}')
print(f'Coeficiente de mortalidade em Hubei {mortality_rate}')
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-2], recovered.sum()[3:-2], deaths.sum()[3:-2]],
            index=['confirmed', 'recovered', 'deaths']
                              ).T
```

```python
grouped_per_day['letality_rate_1'] = grouped_per_day['deaths'] / grouped_per_day['confirmed'] *100
grouped_per_day['letality_rate_2'] = grouped_per_day['deaths'] / (grouped_per_day['deaths'] 
                                                                  + grouped_per_day['recovered']) *100
```

```python
plt.figure(figsize=(10,5))
ax = grouped_per_day['letality_rate_1'].plot()
plt.legend(['Mortes/casos confirmados'])
ax.set_xlabel('Data')
ax.set_ylabel('% letalidade')
```
