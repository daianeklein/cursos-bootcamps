# Alucar - Analisando as vendas

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
alucar = pd.read_csv('dados/alucar.csv')
```

```python
## Verificando a quantidade de linhas e colunas
print('Quantidade de linhas e colunas:' , alucar.shape)
```

```python
## Verificando a quantidade de dados nulos

print('Quantidade de dados nulos: ', alucar.isna().sum().sum())
```

```python
## Verificar o tipo de dados 
alucar.dtypes
```

```python
## Alterar o tipo de dado

alucar['mes'] = pd.to_datetime(alucar['mes'])
alucar.dtypes
```

```python
## Plotando os dados em um grafico

sns.lineplot(x='mes', y='vendas', data=alucar)
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes', y='vendas', data = alucar)
ax.figure.set_size_inches(12, 6)
ax.set_title('Vendas Alucar de 2017 e 2018', loc='left', fontsize=16)
ax.set_xlabel('Meses', fontsize = 14)
ax.set_ylabel('Vendas R$', fontsize = 14)
ax = ax
```

## "Decompondo" as vendas

```python
## Calculando a diferena entre as vendas, ou seja, quanto foi o aumento em cada ano

alucar['aumento'] = alucar['vendas'].diff()
alucar
```

```python
## Plotando as diferencas no grafico

sns.set_palette('Accent')
sns.set_style('darkgrid')
ax = sns.lineplot(x='mes', y='aumento', data = alucar)
ax.figure.set_size_inches(12,6)
ax.set_title('Quantidade vendida por mes', loc='left', fontsize=16)
ax.set_xlabel('Meses', fontsize = 14)
ax.set_ylabel('Vendas', fontsize = 14)
ax = ax
```

```python
## Criando uma funcao com todas as configuracoes do 'plot' do grafico

def plotar (titulo, labelx, labely, x, y, dataset):
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x=x, y=y, data = dataset)
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc='left', fontsize=16)
    ax.set_xlabel(labelx, fontsize = 14)
    ax.set_ylabel(labely, fontsize = 14)
    ax = ax
```

```python
plotar('Aumento das vendas Alucar 2017 e 2018', 'Data', 'Aumento', 'mes', 'aumento', alucar)
```

## Calculando a acelerao

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('Aceleracao das vendas da Alucar de 2017 e 2018', 'Tempo', 'Aceleracao', 'mes', 'aceleracao', dataset = alucar)
```

## "Juntando" os tres graficos

```python
plt.figure(figsize=(16,8))
ax = plt.subplot(3, 1, 1) ## 3 quer dizer que o grafico sera dividido em tres partes. 1 quer dizer em qual plot sera
ax.set_title('Analise da Alucar de 2017 e 2018', fontsize = 18, loc='left')
sns.lineplot(x='mes', y='vendas', data=alucar)

##Separando a nova "parte"
plt.subplot(3, 1, 2)
sns.lineplot(x='mes', y='aumento', data=alucar)

##Separando a terceira "parte"
plt.subplot(3, 1, 3)
sns.lineplot(x='mes', y='aceleracao', data=alucar)

ax= ax
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    plt.figure(figsize=(16,8))
    ax = plt.subplot(3, 1, 1) ## 3 quer dizer que o grafico sera dividido em tres partes. 1 quer dizer em qual plot sera
    ax.set_title(titulo, fontsize = 18, loc='left')
    sns.lineplot(x=x, y=y1, data=dataset)

    ##Separando a nova "parte"
    plt.subplot(3, 1, 2)
    sns.lineplot(x=x, y=y2, data=dataset)

    ##Separando a terceira "parte"
    plt.subplot(3, 1, 3)
    sns.lineplot(x=x, y=y3, data=dataset)

    ax= ax
```

```python
plot_comparacao('mes', 'vendas', 'aumento', 'aceleracao', alucar, 'Analise das vendas da Alucar')
```

## Autocorrelao

```python
## Nivel maximo de correlacao  1
## Nivel minimo de correlacao  -1

## Importando biblioteca que faz essa comparacao

from pandas.plotting import autocorrelation_plot
```

```python

ax = plt.figure(figsize=(10,5))
ax.suptitle('Correlacao das vendas', fontsize= 14, x=0.26, y=0.95)
autocorrelation_plot(alucar['vendas'])
ax = ax
```

```python
## Correlacao do aumento

ax = plt.figure(figsize=(10,5))
ax.suptitle('Correlacao das aumento', fontsize= 14, x=0.26, y=0.95)
autocorrelation_plot(alucar['aumento'][1:])
ax = ax
```

```python
## Correlacoa da aceleracao

ax = plt.figure(figsize=(10,5))
ax.suptitle('Correlacao das aumento', fontsize= 14, x=0.26, y=0.95)
autocorrelation_plot(alucar['aumento'][2:])
ax = ax
```

## Vendas Chocolura (Outubro e Novembro)

```python
vendas_por_dia = pd.read_csv('dados/vendas_por_dia.csv')
vendas_por_dia.head()
```

```python
vendas_por_dia.shape
```

```python
vendas_por_dia.isna().sum().sum()
```

```python
vendas_por_dia.dtypes
```

```python
vendas_por_dia['dia'] = pd.to_datetime(vendas_por_dia['dia'])
vendas_por_dia.dtypes
```

```python
vendas_por_dia['aumento'] = vendas_por_dia['vendas'].diff()
vendas_por_dia['aceleracao'] = vendas_por_dia['aumento'].diff()

vendas_por_dia.head()
```

```python
plot_comparacao('dia', 'vendas', 'aumento', 'aceleracao', vendas_por_dia, 'analise de vendas')
```

```python
## Criando uma coluna do dia da semana

vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
vendas_por_dia.head()
```

```python
## Alterando a coluna dia_da_semana para portugues

vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_tradu = {'Monday' : 'segunda', 'Tuesday' : 'terca', 'Wednesday' : 'quarta', 'Thursday' : 'quinta', 'Friday' : 'sexta', 'Saturday' : 'sabado', 'Sunday' : 'domingo' }
```

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_tradu)
vendas_por_dia.head()
```

```python
## Media de vendas por dia

vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas', 'aumento', 'aceleracao'].mean().round()
```

```python
vendas_agrupadas
```

```python
vendas_por_dia
```

## Cafelura - Analise de vendas

```python
cafelura = pd.read_csv('dados/cafelura.csv')
```

```python
cafelura['mes'] = pd.to_datetime(cafelura['mes'])
```

```python
plotar('Vendas Cafelura', 'Tempo', 'Vendas', 'mes', 'vendas', cafelura)
```

```python
quantidade_de_dias_de_fds = pd.read_csv('dados/dias_final_de_semana.csv')
quantidade_de_dias_de_fds['quantidade_de_dias'].values
```

```python
cafelura['vendas_normalizadas'] = cafelura['vendas'] / quantidade_de_dias_de_fds['quantidade_de_dias'].values
cafelura.head()
```

```python
plotar('Vendas normalizadas', 'Tempo', 'Vendas normalizadas', 'mes', 'vendas_normalizadas', cafelura)
```

```python
plt.figure(figsize=(12,8))
ax = plt.subplot(2,1,1)
ax.set_title('Vendas Cafelura 2017 e 2018', fontsize = 18)

sns.lineplot(x='mes', y='vendas', data=cafelura)
ax = plt.subplot(2,1,2)
ax.set_title('Vendas Normalizadas', fontsize = 18)
sns.lineplot(x='mes', y='vendas_normalizadas', data=cafelura)
ax = ax
```
