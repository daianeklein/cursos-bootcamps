```python
from shapely.geometry import Polygon, Point, LineString, MultiPolygon
```

```python
# Criando um quadrado
# Lembrar do plano cartesiano, X e Y
# Quadrado == 4 pontos

p1 = Polygon([(0, 0), (1,0), (1, 1), (0, 1)])

g = gpd.GeoSeries(p1)
print(g)
print(g.plot())
```

```python
# Criando um triangulo
p2 = Polygon([(0,0), (1,0), (1,1)])
g = gpd.GeoSeries(p2)
print(g)
print(g.plot())
```

```python
# Criando um quadrado II
p3 = Polygon([(2,0), (3,0), (3,1), (2,1)])

g = gpd.GeoSeries(p3)
print(g)
print(g.plot())

```

```python
# LineString
p4 = LineString([(0, 1), (3, 0), (1,1)])

g = gpd.GeoSeries(p4)
print(g)
print(g.plot())
```

```python
# LineString
# Inserindo um ponto
p4 = LineString([(0, 1), (3, 0), (1,1)])
p5 = Point(0.5, 0.5)

g = gpd.GeoSeries(p4)
print(g)
print(g.plot())
```

```python
# Tupla de poligonos
p6 = Polygon([(1,0), (1.5, 0.4), (2,0)])
p7 = Polygon([(1,1), (1.5, 0.6), (2,1)])

p8 = MultiPolygon([p6, p7])

g = gpd.GeoSeries(p8)
print(g)
g.plot()
```
