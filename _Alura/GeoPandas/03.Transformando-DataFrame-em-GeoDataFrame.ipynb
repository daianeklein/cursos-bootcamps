```python
%matplotlib inline

import pandas as pd 
import geopandas as gpd
```

```python
dados = pd.read_table('dados.txt')
dados
```

```python
type(dados) #dataframe
```

```python
from shapely.geometry import Point
```

```python
# x == latitude
# y== longitude

x = zip(dados.Longitude, dados.Latitude)
```

```python
x 
```

```python
list(x)
```

```python
geometry = [Point(x) for x in zip(dados.Longitude, dados.Latitude)]
```

```python
crs = {'proj' : 'latlong', 'ellps' : 'WGS84', 'datum' : 'WGS84', 'no_defs' : True}
geo_dados = gpd.GeoDataFrame(dados, crs = crs, geometry = geometry)
```

```python
geo_dados # criada a coluna 'geometry'
```

```python
type(geo_dados)
```

```python
# exportando

import os 

dir = 'Mapas/RJ_DATASET'
if not os.path.exists(dir):
    os.makedirs(dir)
    
geo_dados.to_file(dir + '/DATASET.shp')
```

```python
geo_dados.plot(figsize=(12, 6))
```
