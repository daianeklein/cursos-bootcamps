# DADOS CENSITARIOS

```python
%matplotlib inline

import pandas as pd 
import geopandas as gpd
```

```python
censo = pd.read_excel('../01.Dados/Censo2010/RJ/EXCEL/Basico_RJ.xls')
```

```python
setor = gpd.read_file('../01.Dados/Mapas/RJ-SETOR/RJ-SETOR.shp')
```

```python
censo.head()
```

```python
setor.head()
```

```python
# Alterando o tipo de dado
censo['Cod_setor'] = censo['Cod_setor'].astype(str) 

censo.dtypes
```

```python
#tipo de dados de cada variavel
print(type(censo))
print(type(setor))
```

```python
setor.shape #tamanho da variavel setor
```

```python
#fazendo merge
setor = pd.merge(setor, censo, left_on='CD_GEOCODI', right_on='Cod_setor', how='left')

#tamanho da variavel
setor.shape 
```

```python
setor.crs 
```

```python
# Salvando arquivo
setor.to_file('../01.Dados/Mapas/RJ-SETOR/RJ-SETOR.shp')
```
