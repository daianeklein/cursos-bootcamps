```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Grafico de dispersao', fontsize = 20, pad = 20)
eixo.set_xlabel('Comprimento da sépala', fontsize = 10)
eixo.set_ylabel('Largura da sépala', fontsize = 10)
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Grafico de dispersao', fontsize = 20, pad = 20)
eixo.set_xlabel('Comprimento da sépala', fontsize = 10)
eixo.set_ylabel('Largura da sépala', fontsize = 10)

## Aumentando a fonte dos eixos
eixo.tick_params(labelsize=12)
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

## Colorindo as 'bolinhas' conforme cada especie

cores = {'Iris-setosa' : 'red', 'Iris-versicolor' : 'blue', 'Iris-virginica' : 'green'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie])


eixo.set_title('Grafico de dispersao', fontsize = 20, pad = 20)
eixo.set_xlabel('Comprimento da sépala', fontsize = 10)
eixo.set_ylabel('Largura da sépala', fontsize = 10)
eixo.legend(cores, fontsize=12)

## Aumentando a fonte dos eixos
eixo.tick_params(labelsize=12)
```

```python
fig = plt.figure(figsize=(10,6))
eixo = fig.add_axes([0,0,1,1])

## Colorindo as 'bolinhas' conforme cada especie

cores = {'Iris-setosa' : 'red', 'Iris-versicolor' : 'blue', 'Iris-virginica' : 'green'}

marcadores = {'Iris-setosa' : 'x', 'Iris-versicolor' : 'o', 'Iris-virginica' : 'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color=cores[especie], marker=marcadores[especie])


eixo.set_title('Grafico de dispersao', fontsize = 20, pad = 20)
eixo.set_xlabel('Comprimento da sépala', fontsize = 10)
eixo.set_ylabel('Largura da sépala', fontsize = 10)
eixo.legend(cores, fontsize=12)

## Aumentando a fonte dos eixos
eixo.tick_params(labelsize=12)
```

```python
fig=plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins = 20, orientation='horizontal')
eixo.set_title('Histograma', fontsize = 12, pad = 10)
eixo.set_xlabel('Comprimento da pétala')
eixo.grid(True)
```

```python
fig=plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins = 20, density=True)
eixo.set_title('Histograma', fontsize = 12, pad = 10)
eixo.set_xlabel('Comprimento da pétala')
eixo.grid(True)
```

```python
fig=plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

##media = μ ---na declaracao de variavel, utiliza-se 'mu'

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 12, pad = 10)
eixo.set_xlabel('Comprimento da pétala')
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu,sigma),
              xy=(4.5, 20), fontsize = 20)
```

```python
fig=plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

##media = μ ---na declaracao de variavel, utiliza-se 'mu'

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 12, pad = 10)
eixo.set_xlabel('Comprimento da pétala')
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu,sigma),
              xy=(4.5, 20), fontsize = 20)

eixo.axvline(mu, color='orange', linestyle='--')
eixo.annotate('média', xy=(mu-1.3, 28), fontsize = 20)
```

```python
fig=plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

##media = μ ---na declaracao de variavel, utiliza-se 'mu'

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 12, pad = 10)
eixo.set_xlabel('Comprimento da pétala')
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$/sigma = {1:.2f}$'.format(mu,sigma),
              xy=(4.5, 20), fontsize = 20)

eixo.axvline(mu, color='orange', linestyle='--')
eixo.annotate('média', xy=(mu-1.3, 28), fontsize = 20)

eixo.axvline(df['comprimento_pétala'].median(), color = 'green', linestyle = '--')
eixo.annotate('mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize = 20)

fig.savefig('histograma_ivg.png', bbox_inches='tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_ivg.png').size
combinada = Image.new("RGB", (3*largura, altura))
intervalo = 0
for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png',
                               'histograma_ivg.png']):
    combinada.paste(imagem, (intervalo, 0))
    intervalo += largura
    
combinada.save('combinada.png')
```
