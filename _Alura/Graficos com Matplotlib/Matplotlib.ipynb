## Primeiras visualizacoes

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
print(matplotlib.__version__)
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'cadetblue')
eixo.set_title('Temperatura no momento', fontsize = 16)
eixo.set_ylabel('Temperatura', fontsize = 12)
eixo.set_xlabel('Data', fontsize = 12)
eixo.legend(['temperatura'], loc = 'lower right', fontsize = 10)

```

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize = 18, pad=20)
eixo.set_xlabel('Data', fontsize  = 10)
eixo.set_ylabel('Temperatura', fontsize = 10)
eixo.legend(['Temperatura'], loc='lower right', fontsize = 15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

eixo.annotate('Maximo', xy=(x1, y1), fontsize = 16)

eixo.axhline(max(df['temperatura']), color = 'red', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'red', linestyle='--')
```

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize = 18, pad=20)
eixo.set_xlabel('Data', fontsize  = 10)
eixo.set_ylabel('Temperatura', fontsize = 10)
eixo.legend(['Temperatura'], loc='lower right', fontsize = 15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura'] - 5)

eixo.annotate('Maximo', xy=(x1, y1), fontsize = 11, xytext = (x2, y2), arrowprops=dict(facecolor='grey'))

eixo.axhline(max(df['temperatura']), color = 'red', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'red', linestyle='--')
```

```python
fig = plt.figure(figsize=(10,5))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize = 18, pad=20)
eixo.set_xlabel('Data', fontsize  = 10)
eixo.set_ylabel('Temperatura', fontsize = 10)
eixo.legend(['Temperatura'], loc='lower right', fontsize = 15)
eixo.grid(True)

## Definindo a seta 'maximo'
x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura'] - 5)

eixo.annotate('Maximo', xy=(x1, y1), fontsize = 11, xytext = (x2, y2), arrowprops=dict(facecolor='grey'))

## Definindo a seta 'minimo'
x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura'] + 5)

eixo.annotate('Minimo', xy =(x1, y1), fontsize = 11, xytext = (x2, y2), arrowprops=dict(facecolor='grey'))


eixo.axhline(max(df['temperatura']), color = 'red', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'red', linestyle='--')
```

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']

temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]

temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))

eixo.bar(indice, temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 20)
eixo.set_xlabel('Dia da semana', fontsize = 12)
eixo.set_ylabel('temperatura Media', fontsize = 12)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'red', 'green', 'blue', 'yellow', 'orange', 'purple']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores)
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 20)
eixo.set_xlabel('Dia da semana', fontsize = 12)
eixo.set_ylabel('temperatura Media', fontsize = 12)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'red', 'green', 'blue', 'yellow', 'orange', 'purple']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor='grey')
eixo.set_title('Temperatura por dia da semana', fontsize = 15, pad = 20)
eixo.set_xlabel('Dia da semana', fontsize = 12)
eixo.set_ylabel('temperatura Media', fontsize = 12)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index)
eixo.set_title('Temperatura por dia da semana', fontsize = 14, pad = 20)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
        autopct='%0.1f%%')

eixo.set_title('Temperatura por dia da semana', fontsize = 14, pad = 20)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
explodir = [0.1, 0, 0, 0, 0, 0, 0.1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
        autopct='%0.1f%%', explode=explodir, shadow=True)

eixo.set_title('Temperatura por dia da semana', fontsize = 14, pad = 20)
```
