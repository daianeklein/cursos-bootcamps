## Importando base de dados

### Fonte: https://s3.amazonaws.com/caelum-online-public/1039-seaborn/material/tips.csv

### Renomeando colunas

```python
## Ver as colunas

dados.columns
```

## Teste de Hipotese

### Hipotese 1: A distribuicao da taxa da gorjeta é a mesma nos dois grupos
### Hipotese 2: A distribuicao da taxa da gorjeta nao é a mesma nos dois grupos

```python
from scipy.stats import ranksums
```

```python
com_sobremesa = gorjetas.query('sobremesa == "Sim"').porcentagem
sem_sobremesa = gorjetas.query('sobremesa == "nao"').porcentagem
```

```python
r = ranksums(com_sobremesa, sem_sobremesa)
```

```python
print('O valor do p-value é: {}'.format(r.pvalue))
```
