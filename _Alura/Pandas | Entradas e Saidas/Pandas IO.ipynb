# Criando a lista de nomes de alunos

## API IBGE

```python
import pandas as pd
```

```python
nomes_f = pd.read_json('https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=200&sexo=f')
nomes_m = pd.read_json('https://servicodados.ibge.gov.br/api/v1/censos/nomes/ranking?qtd=200&sexo=m')
```

```python
print('Quantidade de alunos: ' + str(len(nomes_f) + len(nomes_m)))
```

### Juntando as duas variaveis em uma unica lista

```python
frames = [nomes_f, nomes_m]
frames
```

```python
nomes = pd.concat(frames)['nome'].to_frame()
```

## Incluindo o ID dos alunos

```python
#biblioteca numpy para que seja gerado um numero aleatorio
import numpy as np
np.random.seed(123)
```

```python
total_alunos = len(nomes)
total_alunos
```

```python
nomes['ID_aluno'] = np.random.permutation(total_alunos) + 1
```

```python
nomes.sample(5)
```

```python
dominios = ['@dominiodoemail.com.br', '@servicodoemail.com']
nomes['dominio'] = np.random.choice(dominios, total_alunos)
nomes.sample(5)
```

### Concatenando strings

```python
#nomes.nome.= do dataframe 'nomes', pegando a coluna 'nome'
#str para transformar em str ---- type(nomes) == pandas.core.frame.DataFrame
#Cat para juntar 
#nomes == coluna nomes // dominio == coluna dominio criada no code anterior

nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower()
```

```python
nomes.sample(5)
```

```python
url = 'http://tabela-cursos.herokuapp.com/index.html'
cursos = pd.read_html(url)
##type(cursos)
```

### Transformando uma lista em um DataFrame

```python
#variavel 'cursos' == cursos = cursos[0]
cursos = cursos[0]
```

```python
type(cursos)
```

```python
cursos.columns = ['nome_do_curso']
```

```python
## Segunda forma para mudar o nome da coluna
cursos = cursos.rename(columns={'nome_curso' : 'nome_do_curso'})
```

```python
cursos.head()
```

### Criando ID dos cursos

```python
cursos['id'] = cursos.index + 1
```

```python
cursos.head()
```

```python
cursos = cursos.set_index('id')
cursos.head()
```

## Atribuindo os alunos aos cursos

```python
##Criando a coluna 'matriculas'
## a funcao 'exponencial', neste caso, ira "distribuir" varios cursos para os alunos --> um para muitos
##astype(int) == faz com que a coluna criada 'matricula' seja do tipo int

nomes['matricula'] = np.ceil(np.random.exponential(size = total_alunos) * 1.5 ).astype(int)
nomes.sample(15)
```

```python
nomes.matricula.describe()
```

## Selecionando quais os cursos cada aluno esta matriculado

```python
todas_matriculas = []
x = np.random.rand(20)
prob = x / sum(x)
```

```python
for index, row in nomes.iterrows():
    id = row.ID_aluno
    matriculas = row.matricula
    for i in range(matriculas):
        mat = [id, np.random.choice(cursos.index, p = prob)]
        todas_matriculas.append(mat)

matriculas = pd.DataFrame(todas_matriculas, columns = ['ID_aluno', 'id_curso'])
```

```python
nomes
```

```python
cursos
```

```python
matriculas.head()
```

```python
matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'ID_aluno' : 'quantidade_de_alunos'})
```

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns={'ID_aluno' : 'quantidade_de_alunos'})
```

```python
matriculas_por_curso.head()
```
